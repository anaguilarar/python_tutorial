# Analysis of Land Cover 

Land cover analysis consist in the characterization of different land classes thorugh satellite information. We can explorer the reflectances of each land cover for each satellite bands.


### Data extraction

```python
import rasterio
import geopandas as gpd
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from rasterio.mask import mask
import rasterio.plot


# read in the polygons data
polygons = gpd.read_file("data/polygons.shp")

# read in the image data
imag_data = rasterio.open("data/S2_20170718.tif")
imag_metadata = imag_data.meta.copy()

```

```python
def get_features(gpd_vector, element):
    """Function to parse features from GeoDataFrame in such a manner that rasterio wants them"""
    import json
    jsonFormat = json.loads(gpd_vector.to_json())['features'][element]['geometry']
    return(jsonFormat)
```

```python
## select the land cover of interest
dataframe = pd.DataFrame(polygons)
rowindex = dataframe.loc[dataframe.cover == 'water'].index

## save the spectral data
water_data = []

## extract spectral data using polygon vectors
for i in rowindex:
    jsonfeature = get_features(polygons, i)
    out_img, out_transform = mask(imag_data, [jsonfeature], crop=True, nodata=65535)
    
    water_data.append(out_img)
    
```

```python
def image_histogram_equalization(image):
    """ Image histogram equlisation"""
    # from http://www.janeriksolem.net/2009/06/histogram-equalization-with-python-and.html

    # get image histogram
    image_histogram, bins = np.histogram(image.flatten(), 256, normed=True)
    cdf = image_histogram.cumsum() # cumulative distribution function
    cdf = (255-1) * cdf / cdf[-1] # normalize

    # use linear interpolation of cdf to find new pixel values
    image_equalized = np.interp(image.flatten(), bins[:-1], cdf)
    
    if len(image.shape) == 3:
        return image_equalized.reshape(image.shape).astype('uint8')[:,:,::-1]
    else:
        return image_equalized.reshape(image.shape).astype('uint8')
```

```python
## remove None data
mask_nadata = water_data[0][2,:,:]!=int(imag_metadata['nodata'])

plt.figure(figsize=(10,10))
plt.imshow(image_histogram_equalization(water_data[0][2,:,:]* mask_nadata))
plt.colorbar(orientation="horizontal", pad=0.1)
plt.title('red')
```

```python
def extract_landcoverdata(polygons, label):
    ## select the land cover of interest
    dataframe = pd.DataFrame(polygons)
    rowindex = dataframe.loc[dataframe.cover == label].index
    ## save the spectral data
    land_data = []

    ## extract spectral data using polygon vectors
    for i in rowindex:
        jsonfeature = get_features(polygons, i)
        out_img, out_transform = mask(imag_data, [jsonfeature], crop=True)

        land_data.append(out_img)
        
    
    return land_data
```

### Extract all data from the polygons and save them as pandas

```python

### band spectral names
band_names = ['blue', 'green', 'red' , 'nir', 'swir1']

## organize data as pandas dataframe
df_reflectance = []

for label in np.unique(dataframe.cover):
    
    df_features = []
    ## extract data from polygons
    features_data = extract_landcoverdata(polygons, label)
    for j in range(len(features_data)):
        geomfeature = features_data[j]
        ## transform 
        df = pd.DataFrame([i.ravel() for i in geomfeature])
        df = df.transpose()
        df.columns = band_names
        df['polygon'] = j
        df_features.append(df)
    
    df_perlabel = pd.concat(df_features)
    df_perlabel['landcover'] = label
    df_reflectance.append(df_perlabel)

df_reflectance = pd.concat(df_reflectance) 
df_reflectance.head()
```

### compare land cover refletances



```python
### reshape the original dataframe
## initial shape
print(df_reflectance.head())

## reshape in long direction
df_plot = pd.melt(df_reflectance.drop(['polygon'], axis=1), id_vars=['landcover'])
df_plot.head()
```

```python
import seaborn as sns
import math
plt.figure(figsize=(12,8))

## remove nodata values
df_plot = df_plot.loc[df_plot.value != int(imag_metadata['nodata'])]

### calculate mean values
data_grouped = df_plot.groupby(['landcover', 'variable']).mean().reset_index()

## create plots
plots = []
fig, ax = plt.subplots()
labels = data_grouped.landcover.unique()
for i in range(len(data_grouped.landcover.unique())):
    data_per_cover = data_grouped.loc[data_grouped.landcover == labels[i],:].value.values.tolist()
    coverplot = ax.plot( data_grouped.variable.unique().tolist(),np.array(data_per_cover),
                        label=labels[i] , marker='o')
    
    plots.append(coverplot)

ax.legend(tuple([i[0] for i in plots]), tuple(labels.tolist()),bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### Suppervised Classification with Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize our model with 500 trees
rf = RandomForestClassifier(n_estimators=500, oob_score=True)

## split the data
# take first plygon as validation dataset
training_data = df_reflectance.loc[df_reflectance.polygon !=0]

validation_data = df_reflectance.loc[df_reflectance.polygon ==0]

## train the model
rf = rf.fit(training_data[['blue', 'green', 'red', 'nir', 'swir1']], training_data['landcover'])
```

```python
## compare true data vs predic data
```

```python
validation_data['model_class'] = rf.predict(validation_data[['blue', 'green', 'red', 'nir', 'swir1']])

print(pd.crosstab(validation_data['landcover'], validation_data['model_class'], margins=True))
```

### classify the image

```python
band_values = []
## organice the rasterio as pandas dataframe
for i in range(len(['blue', 'green', 'red', 'nir', 'swir1'])):
    band_values.append(imag_data.read(i+1).ravel())

imageaspandas = pd.DataFrame(np.transpose(np.array(band_values)), columns=['blue', 'green', 'red', 'nir', 'swir1'])
imageaspandas.head()
```

```python
image_class = rf.predict(imageaspandas)
```

```python
ref_data = imag_data.read()
## assign the classified values in the image
reshapedimage_class= image_class.reshape(ref_data.shape[1],ref_data.shape[2])
labels = np.unique(reshapedimage_class)

for i in range(len(labels)):
    reshapedimage_class[reshapedimage_class == labels[i]] = i

reshapedimage_class =reshapedimage_class.astype(np.uint8)
```

```python
from matplotlib.colors import ListedColormap
from matplotlib import colors as mcolors
from matplotlib.patches import Patch


## create random colors
colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
colors = [np.random.choice([i for i in colors.keys()]) for i in range(len(labels))]

## merge color and label as legend patch
legend_patches = [Patch(color=icolor, label=label)
                  for icolor, label in zip(colors, labels)]

cmap = ListedColormap(colors)

fig, ax = plt.subplots(figsize=(12, 8))
ax.imshow(reshapedimage_class, interpolation='none', cmap=cmap)
ax.legend(handles=legend_patches,
         facecolor ="white",
         edgecolor = "white",
         bbox_to_anchor=(1.05, 1), loc=2)
```
